```python
import numpy as np 
import pandas as pd 

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly.express as px

import warnings
warnings.filterwarnings('ignore')

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
train_data = pd.read_csv("./../data/raw/Sign Language MNIST/sign_mnist_train.csv")
test_data = pd.read_csv("./../data/raw/Sign Language MNIST/sign_mnist_test.csv")
```

```python
# Extract the label column from the training dataframe
train_labels = train_data['label']
train_labels.head()  # Display the first few labels

# Create the training set by dropping the 'label' column
train_features = train_data.drop(columns=['label'])
train_features.head()  # Display the first few rows of the feature set
```

```python
# Convert the training feature set to a NumPy array and reshape it to the desired format
X_train = train_features.values.reshape(-1, 28, 28, 1)

# Display the shape of the reshaped array
print(f"Reshaped X_train shape: {X_train.shape}")
```

```python
# Extract the labels from the test dataframe
test_labels = test_data['label']

# Create the test feature set by dropping the 'label' column
X_test = test_data.drop(columns=['label'])

# Display the shape and the first few rows of the test feature set
print(f"Shape of X_test: {X_test.shape}")
X_test.head()
```

```python
from sklearn.preprocessing import LabelBinarizer

# Initialize the LabelBinarizer
label_binarizer = LabelBinarizer()

# Convert training and test labels to one-hot encoded format
y_train = label_binarizer.fit_transform(train_labels)
y_test = label_binarizer.transform(test_labels)

# Reshape the test feature set to the desired format
X_test = X_test.values.reshape(-1, 28, 28, 1)

# Print the shapes of the datasets
print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}, y_test shape: {y_test.shape}")
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Create an ImageDataGenerator for augmenting the training data
train_datagen = ImageDataGenerator(
    rescale=1.0 / 255, 
    rotation_range=0,
    height_shift_range=0.2,
    width_shift_range=0.2,
    shear_range=0,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Normalize the test feature set
X_test = X_test / 255.0

# (Optional) Display confirmation
print("Training data augmentation setup complete. Test data normalized.")
```

```python
# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(8, 8))
fig.suptitle('Preview of Dataset', fontsize=16)

# List of indices, labels, and corresponding letters
preview_data = [
    (0, "label: 3  letter: C"),
    (1, "label: 6  letter: F"),
    (2, "label: 2  letter: B"),
    (4, "label: 13  letter: M")
]

# Loop through the data and display each image with its label
for ax, (index, title) in zip(axes.flatten(), preview_data):
    ax.imshow(X_train[index].reshape(28, 28), cmap='gray')
    ax.set_title(title)
    ax.axis('off')  # Hide axes for a cleaner look

# Show the plot
plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust layout to make room for the title
plt.show()
```

```python
# Plot the frequency of each label
plt.figure(figsize=(10, 6))
sns.countplot(x=train_labels, palette='coolwarm')

# Add title and labels
plt.title("Frequency of Each Label", fontsize=16, fontweight='bold')
plt.xlabel("Labels", fontsize=12)
plt.ylabel("Frequency", fontsize=12)

# Display the plot
plt.show()
```

## Building the CNN Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten

# Initialize the model
model = Sequential()

# Add convolutional and pooling layers
model.add(Conv2D(filters=128, kernel_size=(5, 5), strides=1, padding='same', activation='relu', input_shape=(28, 28, 1)))
model.add(MaxPooling2D(pool_size=(3, 3), strides=2, padding='same'))
model.add(Conv2D(filters=64, kernel_size=(2, 2), strides=1, padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2), strides=2, padding='same'))
model.add(Conv2D(filters=32, kernel_size=(2, 2), strides=1, padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2), strides=2, padding='same'))

# Flatten the output
model.add(Flatten())
```

```python
from tensorflow.keras.layers import Dense, Dropout

# Add a fully connected layer with ReLU activation
model.add(Dense(units=512, activation='relu', name='dense_layer_1'))

# Add a dropout layer to reduce overfitting
model.add(Dropout(rate=0.25, name='dropout_layer'))

# Add the output layer with softmax activation for multi-class classification
model.add(Dense(units=24, activation='softmax', name='output_layer'))

# Display the model summary
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])

# Optional: Display a confirmation message
print("Model compiled successfully with Adam optimizer and categorical crossentropy loss.")
```

```python
# Train the model
history = model.fit(train_datagen.flow(X_train, y_train, batch_size=200), epochs=35, validation_data=(X_test, y_test), shuffle=True, verbose=1)

# Optional: Display a message after training is complete
print("Training completed successfully.")
```

```python
# Evaluate the model on the test data
loss, accuracy = model.evaluate(X_test, y_test)

# Print the model accuracy in percentage format
print(f'Model Accuracy: {accuracy * 100:.2f}%')
```
